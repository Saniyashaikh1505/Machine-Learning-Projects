```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
x = [15,18,22,25,29,32,35,40]
y = ['Serena','Blair','Nate','Chuck','Dan','Jenny','Lily','Rufus']
```

1. Line Plot

```python
plt.plot(x,y)
plt.title("Age of Celebs")
plt.xlabel("Age")
plt.ylabel("Celebrities")
plt.show()
```

2. Bar Chart

```python
plt.bar(x,y)
plt.title("Age of Celebs")
plt.xlabel("Age")
plt.ylabel("Celebrities")
plt.show()
```

3. Histogram

```python
data = np.random.randn(1000)
plt.hist(data , bins = 30)
plt.title('Histogram')
plt.show()
```

4. Scatter Plot

```python
plt.scatter(x,y)
plt.title("Age of Celebs")
plt.xlabel("Age")
plt.ylabel("Celebrities")
plt.show()
```

 Seaborn Examples

```python
df = sns.load_dataset('iris')
df
```

1. Histogram with KDE

```python
sns.histplot(df['sepal_length'] , kde = True)
```

2. Box Plot

```python
sns.boxplot( x = 'species' , y = 'petal_length' , data = df)
plt.title('Box Plot')
plt.show()
```

3. Scatter Plot with hue

```python
sns.scatterplot(x = 'sepal_length' , y = 'sepal_width' , hue = 'species' , data = df)
plt.title('Scatter Plot')
plt.show()
```

4. Correlation Heatmap

```python
sns.heatmap(df.drop('species', axis=1).corr() , annot = True)
plt.title('Correlation Heatmap')
plt.show()
```
