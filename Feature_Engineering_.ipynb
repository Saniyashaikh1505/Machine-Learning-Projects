**Steps in Feature Engineering**

Step 1- Data Cleaning - Handling missing values, handling outliers and handling inconsistencies

Step 2 - Feature Transformation (Feature Encoding and Feature Scaling)
        -> Feature Encoding (Label Encoding, One-Hot Encoding, Freaquncy Encoding, Target Encoding)
        -> Feature Scaling (Normalization , Standarization)

Step 3 - Feature Extraction

Step 4 - Feature Selection

```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('/content/drive/MyDrive/Data Analysis Portfolio - Saniya Shaikh/Machine Learning Projects - Python/Python Data File/employee_promotion.csv')
dataset.head()
```

## **Feature Encoding**

```python
from sklearn.preprocessing import LabelEncoder

df= dataset.copy()

LabelEncoder = LabelEncoder()
df['Promotion_Encoded'] = LabelEncoder.fit_transform(df['Promotion'])
df.head()

df['Education'].unique()
```

```python
# Label Encoding with order - Education
Education_dict = {'High School' : 0, "Bachelor's" : 1, "Master's" : 2, "PhD" : 3}
df['Education_Encoded']= df['Education'].map(Education_dict)
df.head()
```

```python
#One-Hot Encoding - City
df_encoded = pd.get_dummies(df, columns=['City'])
df_encoded.head()
```

```python
#Frequency Encoding - Job Title
job_title_counts = df['JobTitle'].value_counts()
df['JobTitle_Encoded'] = df['JobTitle'].map(job_title_counts)
df.head()
```

# **Feature Scaling**

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler

df_scale = dataset.copy()

#Standard Scaler
standard_scaler = StandardScaler()
df_scale['Standard_Salary'] = standard_scaler.fit_transform(df_scale[['Salary']])

#Min-Max Scaler
min_max_scaler = MinMaxScaler()
df_scale['Min_Max_Salary'] = min_max_scaler.fit_transform(df_scale[['Salary']])

df_scale.head()
```

# **Feature Extraction**

```python
data_cust = pd.read_csv('/content/drive/MyDrive/Data Analysis Portfolio - Saniya Shaikh/Machine Learning Projects - Python/Python Data File/retail_customer.csv')
data_cust.head()
```

```python
#binning age into categories
bins = [10,30,50,70] #Age ranges
labels = ['Young','Mid','Senior']
data_cust['age_group']=pd.cut(data_cust['age'],bins = bins,labels = labels)
data_cust[['age','age_group']].head(20)
```

```python
#converting date column to datetime
data_cust['Signup_date']=pd.to_datetime(data_cust['signup_date'])
data_cust['last_purchase_date']=pd.to_datetime(data_cust['last_purchase_date'])
data_cust.dtypes
```
